```python
# Import the required libraries for analyzing the data on the
# SIRS Model. Four plots are shwon here: the contour on the p1-p3 
# plane of the mean value of the infection rate, the contours on 
# the variance rate
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Read the required data and store it as numpy arrays. Do this using
# the pandas library
df = pd.read_csv("MeanData.csv")
print(df)
```

```python
# Print the contours for the mean values of the infection rates. 
# Do this on imshow using matplotlib function
means = df["Mean"].to_numpy()
means = means.reshape(int(np.sqrt(len(means))), int(np.sqrt(len(means))))

# Now make the plot
plt.figure(figsize=(10,10))
plt.contourf(means, origin='lower', extent=[0,1,0,1], cmap='hot')
plt.colorbar()
plt.title("Mean rate of infection")
plt.xlabel("p1")
plt.ylabel("p3")
```

```python
# Now do the same plots for the variance values in the infection rates
df0 = pd.read_csv("VarianceData.csv")
print(df0)
```

```python
# Now do the contour plot for the variance in the infection rates on 
# the range of probabilities p1 - p3 between 0 and 1
variances = df0["Variance"].to_numpy()
variances = variances.reshape(int(np.sqrt(len(variances))), int(np.sqrt(len(variances))))

# Now make the plot
plt.figure(figsize=(10,10))
plt.contourf(variances, origin='lower', extent=[0,1,0,1], cmap='hot')
plt.colorbar()
plt.title("Variance in the infection rates")
plt.xlabel("p1")
plt.ylabel("p3")
```

```python
# Perform now the data plots for the region of p2 = p3 = 0.5 and
# p1 in the range (0.2, 0.5) with resolution of 0.01 in probability
# units.
df1 = pd.read_csv("InfectedPeople_10000sweeps.csv")
print(df1)
```

```python
P1 = df1["Probabilities_p1 "].to_numpy()
Y = df1["Variances"].to_numpy()
Erry = df1["Error on variance"].to_numpy()
```

```python
plt.figure(figsize=(10,10))
sns.set_theme()
plt.errorbar(P1, Y, yerr=Erry, color='r', fmt='o', markersize=3)
plt.title("Variance vs p1 probability range")
plt.xlabel("p1")
plt.ylabel("Variance")
plt.show()
```

```python
# For the final part, analyze the role of immunity in disease spread. 
# The following assumption has been conducted: p1 = p2 = p3 = 0.5.
# Immunization fractions from 0 to 0.4 have been taken and analysed

# Read the data from the immunization part
df2 = pd.read_csv("Immunity.csv")
print(df2)
```

```python
# Now conduct the plot on matplotlib with seaborn theme
xs = df2["Immunity fraction "].to_numpy()
ys = df2["Infection rate"].to_numpy()
errs = df2["Errors on infection rate"].to_numpy()
```

```python
# Number of measurements for each value
no_meas = 5

# Make now the required plots
plt.figure(figsize=(10,10))
sns.set_theme()
plt.errorbar(xs, ys, yerr=errs / np.sqrt(no_meas), color='r', fmt='o', markersize=3)
plt.title("Infection average rate vs Immunity fraction")
plt.xlabel("Immunity fraction")
plt.ylabel("Average infection rate")
plt.show()
```
